```python
class Product:
    def __init__(self, name,price,quantity):
        self.name= name
        self.price= price
        self.quantity= quantity
        
    def update_quantity(self,new_quantity):
        self.quantity = new_quantity
        
    def total_price(self):
        return self.price * self.quantity

        
class Inventory:
    
    def __init__(self):
        self.products = {}
      
    def add_product(self,product):
        self.products[product.name]=product
        
    def update_product(self,new_product, product):
        self.products[new_product] = self.products.pop(product.name)
        
    def remove_product(self,product_name):
        del self.products[product_name]
        
    def calculate_total_products_price(self):
        total_products_price = 0
        for product in self.products.values():
            total_products_price += product.total_price()
        return total_products_price
        
#Product objects
prod1 = Product("iphone", 1000, 3)
prod2 = Product("laptop", 2000, 2)
prod3 = Product("printer", 200, 5)

#total price function
print(f" The Product total price is {prod1.total_price()}")
    #output: The Product total price is 3000

#update quantity function
prod1.update_quantity(5)
print(f" The Product total price after updating is {prod1.total_price()}")
    #output: The Product total price after updating is 5000
    

#Inventory object
inv = Inventory()

#add product function
inv.add_product(prod1)
inv.add_product(prod2)
inv.add_product(prod3)
print(inv.products.keys())
    #output:dict_keys(['iphone', 'laptop', 'printer'])
    
#update product function
prod4= Product("tablet", 50, 2)
inv.update_product("tablet",prod3)
print(inv.products.keys())
    #output:dict_keys(['iphone', 'laptop', 'tablet'])

#remove product function
inv.remove_product("tablet")
print(inv.products.keys())
    #output:dict_keys(['iphone', 'laptop'])

    
# calculate total price function
total_price = inv.calculate_total_products_price()
print(f"The Total price of Products is : {total_price}")
    #outout:The Total price of Products is : 9000

```
